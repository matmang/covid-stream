```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 감성 분석 완료된 CSV 파일 로드
df = pd.read_csv("covid19_tweets_with_sentiment.csv")
len(df)
```

```python
# 2. 감성 레이블별 개수 집계
sentiment_counts = df['sentiment'].value_counts()
total_counts = sentiment_counts.sum()
sentiment_percentages = (sentiment_counts / total_counts) * 100

print(sentiment_percentages)
# 3. 퍼센테이지 막대 그래프 그리기
plt.figure(figsize=(8,6))
colors = {'positive': 'green', 'neutral': 'blue', 'negative': 'red'}

plt.bar(sentiment_percentages.index, sentiment_percentages.values, color=[colors.get(s) for s in sentiment_percentages.index])

plt.title("Sentiment Distribution in Tweets (Whole Dataset)")
plt.xlabel("Sentiment")
plt.ylabel("Percentage (%)")
plt.grid(axis='y')  # y축 기준 격자 추가
plt.tight_layout()
plt.show()
```
